```python
import math
import numpy as np
import matplotlib.pyplot as plt
import decimal

a = 0.041
m = int(input())
n = int(input())
t = 1.0 / float(m)
h = 1.0 / float(n)
g_coef = 2 * (h ** 2 / (a * t) - 1)
f_coef = (2 * (h ** 2)) / a
d_coef = 2 * (h ** 2 / (a * t) + 1)

def uOT(t):
    return -3 * t ** 3


def uXO(x):
    return 2 * x ** 4 - 2 * x


def uOneT(t):
    return 2 - 3 * t ** 3 + 3 * t ** 2 - 2 * math.exp(t)


def foo(x, t):
    return -9 * t ** 2 + 6 * t * x - 2 * x * math.exp(t) - 24 * a * x ** 2


def check(x, t):
    return 2 * x ** 4 - 3 * t ** 3 + 3 * x * t ** 2 - 2 * x * math.exp(t)

U = [[0] * (m+1) for i in range(n+1)]
for i in range(n+1):
    U[i][0] = uXO(h * i)

for i in range(1, m+1):
    U[0][i] = uOT(t * i)
    U[n][i] = uOneT(t * i)

d = [0] * n
for i in range(1, n):
    d[i] = 1 / (d_coef - d[i-1])

g = [0] * n
for i in range(m):
    for j in range(1, n):
        g[j] = U[j-1][i] + g_coef * U[j][i] + U[j+1][i] + f_coef * foo(h * j, t * (i + 0.5))
    for j in range(1, n):
        U[j][i+1] = d[j] * (g[j] + U[j-1][i+1])
    for j in range(n-1, 0, -1):
        U[j][i+1] += d[j] * U[j+1][i+1]

dist = 0.0
ch = [[0] * (m+1) for i in range(n+1)]
for i in range(n+1):
    for j in range(m+1):
        ch[i][j] = check(h * i, j * t)
        if(abs(ch[i][j] - U[i][j]) > dist):
            dist = abs(ch[i][j] - U[i][j])
print(str(decimal.Decimal(dist)))
tArr = [t * i for i in range(m+1)]
hArr = [h * i for i in range(n+1)]
print()
#for i in range(m+1):
#    print("Solution:", end=" ")
#    for j in range(n+1):
#        print(ch[j][i], end=" ")
#    print()
#    print("Approximation:", end=" ")
#    for j in range(m+1):
#        print(U[j][i], end=" ")
#    print('\n')
```
