```python
import math
import matplotlib.pyplot as plt
a = 0.29
m = int(input())
n = int(input())
t = 1.0 / m
h = 1.0 / n
c = (-a * t) / h
b = 1 + (a * t) / h


def uXO(x):
    return 0.027 + math.cos(2 * x)


def uOT(t):
    return -(t - 0.3) ** 3 + 1


def f(x, t):
    return -3 * (t - 0.3) ** 2 + 2 * math.pi * x - 2 * a * math.sin(2*x) + 2 * a * math.pi * t


def check(x, t):
    return -(t - 0.3) ** 3 + math.cos(2 * x) + 2 * math.pi * t * x


U = [[0] * (m+1) for i in range(n+1)]
ch = [[0] * (m+1) for i in range(n+1)]
dist = 0.0
U[0][0] = uXO(h * 0)

for l in range(n + 1):
    ch[l][0] = check(l * h, 0)

for i in range(1, n + 1):
    U[i][0] = uXO(h * i)
    U[0][i] = uOT(t * i)

for i in range(1, m + 1):

    d = []
    d.append(t * f(0, (i + 1) * t) + uOT(i * t))
    for j in range(1, n + 1):
        d.append(t * f(j * h, (i + 1) * t) + U[j][i-1])
        U[j][i] = ((d[j] - c * U[j-1][i]) / b)

    for l in range(n + 1):
        ch[l][i] = check(l * h, i * t)
        if abs(ch[l][i] - U[l][i]) > dist:
            dist = abs(ch[l][i] - U[l][i])

print(dist)
tArr = [t * i for i in range(m+1)]
hArr = [h * i for i in range(n+1)]
#for i in range(m+1):
#    print("Solution:", end=" ")
#    for j in range(n+1):
#        print(ch[j][i], end=" ")
#    print()
#    print("Approximation:", end=" ")
#    for j in range(m+1):
#        print(U[j][i], end=" ")
#    print('\n')
```

```python
# x = 0.5
plt.figure(figsize = (15, 15))
plt.plot(tArr, U[round(n/2)], '--', tArr, ch[round(n/2)], ':', linewidth = 3)
plt.show()
```
